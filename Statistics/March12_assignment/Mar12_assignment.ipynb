# Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
# we have
CI = 0.95
sample_mean = 50
sample_std = 5
alpha = 1-0.95
n = 50  #assuming
dof = n-1


import scipy.stats as stat
import math
## Now,
t_value = stat.t.ppf(1-alpha/2,dof)

## Putting values in formula for upper bound and lower bound
upper_CI = sample_mean + t_value*(sample_std/math.sqrt(n))
lower_CI = sample_mean - t_value*(sample_std/math.sqrt(n))

print(f"I am 95% sure that my confidence interval is between {upper_CI} and {lower_CI}")
```

# Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare
from scipy.stats import chi2

# observed frequencies
obs_freq = np.array([10, 20, 15, 5, 5, 15])

# expected frequencies
exp_freq = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(obs_freq)

# Null and alternate hypothesis
null_hypothesis = "The distribution of colors of M&Ms in the bag matches the expected distribution."
alternate_hypothesis = "The distribution of colors of M&Ms in the bag does not match the expected distribution."

# perform chi-square goodness of fit test
test_statistic, p_value = chisquare(obs_freq, exp_freq)

# print results
print("Test Chi2 statistic:", test_statistic)
print("p-value:", p_value)

# Critical chi2 value
df = len(obs_freq)-1
alpha = 0.05
chi2_crit = chi2.ppf(1-alpha,df)
print(f"Critical Chi Square value for {alpha} significance level for {df} degrees of freedom is : {chi2_crit:.4f}")

# compare p-value with significance level
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis")
    print(f"Conculsion : {alternate_hypothesis}")
else:
    print("Fail to reject null hypothesis")
    print(f"Conclusion : {null_hypothesis}")
```

```python
import pandas as pd
import scipy.stats as stat
import numpy as np

# firstly we convert this data into dataframe

obs_data = pd.DataFrame({'Group A':[20,10,15],'Group B':[15,25,20]})
obs_data.index = (['Outcome 1','Outcome 2','Outcome 3'])

Null_hypothesis = 'The outcomes are independent of the groups'
Alternate_hypothesis = 'The outcomes are dependent on the groups'

# Now chi stat value and p-value
chi_stat , p_value , dof , expected_data = stat.chi2_contingency(obs_data)

# Now making expected data
expected_df = pd.DataFrame(expected_data)
expected_df.columns = ['Group A',	'Group B']
expected_df.index = (['Outcome 1','Outcome 2','Outcome 3'])

# Now just comparing 

if p_value < alpha:
    print("REJECT NULL")
    print('CONCLUSION: ',Alternate_hypothesis)
else:
    print("ACCEPT NULL")
    print('CONCLUSION: ',Null_hypothesis)
```

# Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
import scipy.stats as stats
import math

# we have
n = 500  # sample size
x = 60   # Number of individuals who smoke in the sample
CI = 0.95
alpha = 1 - 0.95
dof = n -1
# Proportion 
p = x/n

# Calculating standard error.
standard_error = math.sqrt(p * (1 - p) / n)


# Calculating the t-value 
t_value = stats.t.ppf(1-alpha/2, dof)


# Calculating margin of errors
moe = t_value * standard_error

# Calculate the lower and upper bounds of the confidence interval
lower_CI = p - moe
upper_CI = p + moe

print(f"I 95% confident that people who smoke are lie between {lower_CI*100}% and {upper_CI*100}% this proportion")
```

# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

```python
import scipy.stats as stat
import math

# we have 
CI = 0.90
sample_mean = 75
sample_std = 12
alpha = 1 - 0.90
n = 30 # assuming
dof = n-1

# using t-test confidence interval formula

### CI = X̄ ± t*(s/√n)

# calculating t_critical
t_critical = stat.t.ppf(1-alpha/2,dof)

# putting in formula for upper and lower CI
upper_CI = sample_mean + t_critical*(sample_std/math.sqrt(n))
lower_CI =  sample_mean - t_critical*(sample_std/math.sqrt(n))

print(f"I am 90% confident that the confidence interval are between {lower_CI} and {upper_CI}")
```

# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Set the degrees of freedom
df = 10

# Set the x-axis range
x = np.linspace(0, 30, 500)

# Compute the chi-square distribution
y = chi2.pdf(x, df)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y, label='Chi-square Distribution')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(15, 30, 100)
y_fill = chi2.pdf(x_fill, df)
ax.fill_between(x_fill, y_fill, color='orange', alpha=0.5)

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('Density')
ax.set_title(f'Chi-square Distribution with {df} Degrees of Freedom')

# Display the plot
plt.show()
```

```python
# we have
alpha = 0.05
observed_data = [45,55]
expected_data = [50,50]
dof = len(expected_data)-1

Null_hypothesis = 'The coin is fair'
Alternate_hypothesis = 'The coin is biased towards tails'

# Now
import scipy.stats as stat
import math
import numpy as np

# calculating chi stats and p-value
chi_stat , p_value = stat.chisquare(observed_data,expected_data)

# Calculating t_critical 
chi_critical = stat.chi2.ppf(1-alpha , dof)

if chi_stat > chi_critical:
    print("Reject the Null Hupothesis")
    print('CONCLUSION:',Alternate_hypothesis)
else:
    print('Accept the Null hypothesis')
    print('CONCLUSION:',Null_hypothesis)
```

# Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

```python
import math
import scipy.stats as stat

# we have
n = 30
sample_mean = 72
sample_std = 10
alpha = 0.05
dof = n-1

Null_hypothesis = "The population mean is equal to 70."
Alternative_hypothesis = "The population mean is significantly different from 70."

# Calculating the t-statistic
t_stat = (sample_mean - 70) / (sample_std / math.sqrt(n))

# Now p_value
p_value = (1-stat.t.cdf(t_stat,dof))*2

# Comparing
if p_value < alpha:
    print("REJECT NULL")
    print('CONCLUSION: ',Alternative_hypothesis)
else:
    print("ACCEPT NULL")
    print('CONCLUSION: ',Null_hypothesis)
```
